# OpenSea NFT API
https://docs.opensea.io/reference/api-overview

## Asset Model

**Asset** - unique digital item

**Trait** - special properties on the item

**Asset contract** - data about the contract itself

---

## Event Model

**Asset event** - represent state changes that occur for assets, i.e. putting them on sale, bidding on them, selling, cancelling sales, transferring them and more.

---

## Account Model

**Account** Represents wallet addresses and associated usernames (if user entered one on OpenSea.)

---

## Collection Model

**Collection** - used to represent all the assets in a single or multiple contract addresses and help users group items from the same creator. 

---

## Bundles

- Groups of items for sale on OpenSea
- You can buy them all at once in one transaction, and you can create them without any transactions or gas, as long as you've already approved the assets inside.
