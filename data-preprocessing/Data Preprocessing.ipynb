### Initialize new columns in the Dataframe

```python
# Background
df['value_background']='-'
df['display_type_background']='-'
df['max_value_background']='-'
df['trait_count_background']='-'
df['order_background']='-'

# Clothes
df['value_clothes']='-'
df['display_type_clothes']='-'
df['max_value_clothes']='-'
df['trait_count_clothes']='-'
df['order_clothes']='-'

# Earring
df['value_earring']='-'
df['display_type_earring']='-'
df['max_value_earring']='-'
df['trait_count_earring']='-'
df['order_earring']='-'

# Eyes
df['value_eyes']='-'
df['display_type_eyes']='-'
df['max_value_eyes']='-'
df['trait_count_eyes']='-'
df['order_eyes']='-'

# Fur
df['value_fur']='-'
df['display_type_fur']='-'
df['max_value_fur']='-'
df['trait_count_fur']='-'
df['order_fur']='-'

# Hat
df['value_hat']='-'
df['display_type_hat']='-'
df['max_value_hat']='-'
df['trait_count_hat']='-'
df['order_hat']='-'

# Mouth
df['value_mouth']='-'
df['display_type_mouth']='-'
df['max_value_mouth']='-'
df['trait_count_mouth']='-'
df['order_mouth']='-'
```

### Append Trait values into the columns

```python
for index, row in df.iterrows():
#     df['value_background'] = ast.literal_eval(df['traits'])[index]['value']
    base = ast.literal_eval(df.iloc[index]['traits'])
    for i in range(len(base)):
        if base[i]['trait_type']=='Background':
            df.iloc[index, df.columns.get_loc('value_background')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_background')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_background')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_background')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_background')]=base[i]['order']
            
        elif base[i]['trait_type']=='Clothes':
            df.iloc[index, df.columns.get_loc('value_clothes')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_clothes')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_clothes')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_clothes')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_clothes')]=base[i]['order']

        elif base[i]['trait_type']=='Earring':
            df.iloc[index, df.columns.get_loc('value_earring')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_earring')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_earring')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_earring')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_earring')]=base[i]['order']
            
        elif base[i]['trait_type']=='Eyes':
            df.iloc[index, df.columns.get_loc('value_eyes')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_eyes')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_eyes')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_eyes')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_eyes')]=base[i]['order']
            
        elif base[i]['trait_type']=='Fur':
            df.iloc[index, df.columns.get_loc('value_fur')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_fur')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_fur')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_fur')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_fur')]=base[i]['order']
            
        elif base[i]['trait_type']=='Hat':
            df.iloc[index, df.columns.get_loc('value_hat')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_hat')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_hat')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_hat')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_hat')]=base[i]['order']
            
        elif base[i]['trait_type']=='Mouth':
            df.iloc[index, df.columns.get_loc('value_mouth')]=base[i]['value']
            df.iloc[index, df.columns.get_loc('display_type_mouth')]=base[i]['display_type']
            df.iloc[index, df.columns.get_loc('max_value_mouth')]=base[i]['max_value']
            df.iloc[index, df.columns.get_loc('trait_count_mouth')]=base[i]['trait_count']
            df.iloc[index, df.columns.get_loc('order_mouth')]=base[i]['order']
            
        elif base[i]['trait_type']=='Name':
            # This column isn't important - only 5 assets have this data
            continue
```
